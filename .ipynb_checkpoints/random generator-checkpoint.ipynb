```python
# Creation of random distribution in 3D, same dimensions.

x=np.array(range(0,100))

import numpy as np

n = 32700

x1, x2 = 0, 50000
y1, y2 = 0, 50000
z1, z2 = 0, 50000

x = (x2 - x1)*np.random.rand(n) + x1
y = (y2 - y1)*np.random.rand(n) + y1
z = (z2 - z1)*np.random.rand(n) + z1

fig = plt.figure(dpi=200)
ax = plt.axes(projection='3d')
plot = ax.scatter3D(x,y,z,'k.',s=0.5,alpha=0.1)
plt.show()
```
